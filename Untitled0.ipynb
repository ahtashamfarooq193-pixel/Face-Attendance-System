```python
import face_recognition
import pickle

known_encodings = []
known_names = []

import os
for file in os.listdir("/content/dataset"):
    if file.endswith((".jpg", ".jpeg", ".png")):
        img = face_recognition.load_image_file(f"/content/dataset/{file}")
        encs = face_recognition.face_encodings(img)
        if encs:
            known_encodings.append(encs[0])
            name = file.split(".")[0]
            known_names.append(name)
            print(f"✅ {name} encode ho gaya!")
        else:
            print(f"⚠️ {file} mein chehra nahi mila!")

with open("/content/encodings.pkl", "wb") as f:
    pickle.dump({"encodings": known_encodings, "names": known_names}, f)

print("\n🎉 Sab encodings save ho gayi!")
```

```python
import face_recognition
import numpy as np
import pickle
import gspread
from oauth2client.service_account import ServiceAccountCredentials
from datetime import datetime
from PIL import Image
import io
import ipywidgets as widgets
from IPython.display import display, clear_output

# Sheet connect
scope = ["https://spreadsheets.google.com/feeds",
         "https://www.googleapis.com/auth/drive"]
creds = ServiceAccountCredentials.from_json_keyfile_name("credentials.json", scope)
client = gspread.authorize(creds)
sheet = client.open("Attandance sheet").sheet1

# Encodings load
with open("/content/encodings.pkl", "rb") as f:
    data = pickle.load(f)
print(f"✅ {len(data['names'])} log load hue: {data['names']}")
print("\n📸 Ek group photo upload karo — sab ki attendance mark ho jaayegi!\n")

# Attendance function
already_marked = set()
def mark_attendance(name):
    if name in already_marked:
        print(f"⚠️ {name} pehle se mark hai!")
        return
    now = datetime.now()
    sheet.append_row([name, now.strftime("%Y-%m-%d"), now.strftime("%H:%M:%S")])
    already_marked.add(name)
    print(f"✅ {name} — {now.strftime('%H:%M:%S')} — Sheet mein mark ho gaya!")

# Upload button
uploader = widgets.FileUpload(accept='image/*', multiple=False, description='📸 Photo Upload Karo')
out = widgets.Output()
display(uploader, out)

def on_upload(change):
    with out:
        clear_output()
        print("🔍 Faces dhundh raha hoon...\n")

        content = list(uploader.value.values())[0]['content']
        img = Image.open(io.BytesIO(content))
        frame = np.array(img)

        face_locs = face_recognition.face_locations(frame)
        face_encs = face_recognition.face_encodings(frame, face_locs)

        if not face_encs:
            print("😕 Koi chehra nazar nahi aaya — clear photo upload karo!")
            return

        print(f"👥 {len(face_encs)} chehra/chehre mile!\n")

        recognized = []
        unknown = 0

        for enc in face_encs:
            matches = face_recognition.compare_faces(data["encodings"], enc, tolerance=0.5)
            name = "Unknown"
            if True in matches:
                name = data["names"][matches.index(True)]

            if name == "Unknown":
                unknown += 1
            else:
                recognized.append(name)
                mark_attendance(name)

        print(f"\n📊 Summary:")
        print(f"✅ Pehchane gaye: {recognized}")
        if unknown > 0:
            print(f"❓ Nahi pehchane gaye: {unknown} log")
        print(f"\n🎉 Sheet mein mark ho gaye: {list(already_marked)}")

uploader.observe(on_upload, names='value')
```
